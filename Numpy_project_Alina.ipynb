```python
# Download image
! gdown --id 1kQ_UMImy9xM4ml3ci0SGA2CFLtIgpZ1E

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


# Image cropping
def img_cropping(img):

    height, width, dimension = img.shape
    strip_width = int(width/50)
    strips = []

    for x in range(50):
        start_colum = x * strip_width
        end_colum = start_colum + strip_width
        strip = img[:, start_colum:end_colum, :]
        strips.append(strip)
    return strips

# Connect the images
def join(img1, img2, axis):
    join_img = np.concatenate((img1, img2), axis)
    return join_img

# Info about image
def img_info(img):
    return plt.imshow(img), print(img.shape)
```

```python
# Convert image to Numpy array
bird_img = Image.open("bird.jpg")
print(type(bird_img))

np_bird_img = np.asarray(bird_img)
print(type(np_bird_img))

plt.imshow(np_bird_img)
np_bird_img.size, np_bird_img.shape
```

```python
# Image cropping: def img_cropping()

new_img1 = np.concatenate(img_cropping(np_bird_img)[0::2], axis=1)
new_img2 = np.concatenate(img_cropping(np_bird_img)[1::2], axis=1)

img_info(new_img1)
plt.show()
img_info(new_img2)

```

```python
# Connect the images

join_img = join(new_img1,new_img2, 1)

img_info(join_img)
```

```python
# Save image to .jpg format

new_img_2bird = Image.fromarray(join_img)
new_img_2bird.save('new_img_2bird.jpg')
```

```python
# Rotate image
bird2 = Image.open('new_img_2bird.jpg')
ro_bird2 = bird2.rotate(90, expand=1)

print(type(ro_bird2))

# Convert image to Numpy array
np_ro_bird2 = np.asarray(ro_bird2)
print(type(np_ro_bird2))

img_info(np_ro_bird2)
```

```python
# Image cropping: def img_cropping()
new_img3 = np.concatenate(img_cropping(np_ro_bird2)[0::2], axis=1)
new_img4 = np.concatenate(img_cropping(np_ro_bird2)[1::2], axis=1)

img_info(new_img3)
plt.show()
img_info(new_img4)
```

```python
# Connect the images
join_img = join(new_img3,new_img4, 0)

img_info(join_img)
```

```python
# Save image to .jpg format
new_img_4bird = Image.fromarray(join_img)
new_img_4bird.save('new_img_4bird.jpg')

# Rotate image
bird4 = Image.open('new_img_4bird.jpg')
ro_bird4 = bird4.rotate(-90, expand=1)
ro_bird4.save('Bird_4.jpg')

plt.imshow(ro_bird4)

```
